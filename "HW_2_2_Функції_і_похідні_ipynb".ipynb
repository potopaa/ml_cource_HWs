### Домашні завдання на відпрацювання поняття функції та її похідної

1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.



```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

x = sp.symbols('x')
f = 10 * x - 5
f_derivative = sp.diff(f, x)

x_values = np.linspace(0, 10, 400)

f_values = [f.subs(x, value) for value in x_values]
f_derivative_values = [f_derivative.subs(x, value) for value in x_values]

plt.figure(figsize=(10, 6))
plt.plot(x_values, f_values, label='f(x) = 10x - 5', color='blue')
plt.plot(x_values, f_derivative_values, label="f'(x) = 10", color='red')
plt.title('Графік функції та її похідної')
plt.xlabel('Кількість годин навчання (x)')
plt.ylabel('Оцінка')
plt.legend()
plt.grid(True)
plt.show()
```

З графіку видно, що f(x) - лінійна функція з нахилом 10, що означає, що оцінка зростає на 10 одиниць за кожну додаткову годину навчання. А її похідна f'(x) - є константою, що показує постійну швидкість зміни оцінки, яка дорівнює 10 одиницям за годину навчання.

2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



```python
x = sp.symbols('x')

f1 = sp.sin(x)
f1_derivative_ = sp.diff(f1, x)

x_values_sin = np.linspace(-2 * np.pi, 2 * np.pi, 400)

f1_values = [f1.subs(x, value) for value in x_values_sin]
f1_derivative_values = [f1_derivative_.subs(x, value) for value in x_values_sin]

plt.figure(figsize=(12, 6))
plt.plot(x_values_sin, f1_values, label='f(x) = sin(x)', color='blue')
plt.plot(x_values_sin, f1_derivative_values, label="f'(x) = cos(x)", color='red')
plt.title('Графік функції та її похідної: sin(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
x = sp.symbols('x')

f2 = sp.ln(x)
f2_derivative_ = sp.diff(f2, x)

x_values_ln = np.linspace(0.1, 10, 400)

f2_values = [f2.subs(x, value) for value in x_values_ln]
f2_derivative_values = [f2_derivative_.subs(x, value) for value in x_values_ln]

plt.figure(figsize=(12, 6))
plt.plot(x_values_ln, f2_values, label='f(x) = ln(x)', color='blue')
plt.plot(x_values_ln, f2_derivative_values, label="f'(x) = 1/x", color='red')
plt.title('Графік функції та її похідної: ln(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```


3. Загальні витрати компанії від обсягу виробництва задані функцією $ f(x) = x^2 - 5x + 1 $, де $ x $ — кількість вироблених одиниць. Обчисліть похідну цієї функції з допомогою SymPy та візуалізуйте значення функції і її похідної на відрізку від 0 до 100.

```python
x = sp.symbols('x')

f = x**2 - 5*x + 1
f_derivative = sp.diff(f, x)

x_values = np.linspace(0, 100, 400)

f_values = [f.subs(x, value) for value in x_values]
f_derivative_values = [f_derivative.subs(x, value) for value in x_values]

plt.figure(figsize=(12, 6))
plt.plot(x_values, f_values, label='f(x) = x^2 - 5x + 1', color='blue')
plt.plot(x_values, f_derivative_values, label="f'(x) = 2x - 5", color='red')
plt.title('Графік функції та її похідної')
plt.xlabel('Кількість вироблених одиниць (x)')
plt.ylabel('Значення функції')
plt.legend()
plt.grid(True)
plt.show()
```
